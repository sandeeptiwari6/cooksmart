```python
import pandas as pd
```

```python
data = pd.read_csv("data/core-data_recipe.csv")
```

```python
data.ingredients[3].split("^")
```

```python
bag_of_ingredients ={}
for i in range(len(data.ingredients)):
    for ingredient in data.ingredients[i].split("^"):
        if ingredient not in bag_of_ingredients:
            bag_of_ingredients[ingredient]=0
        bag_of_ingredients[ingredient]+=1
```

```python
{k: v for k, v in sorted(bag_of_ingredients.items(), key=lambda item: -item[1])}
```

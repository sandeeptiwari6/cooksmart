```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import pyLDAvis
import pyLDAvis.sklearn
import numpy as np
```

```python
data = pd.read_csv("../data/cleaned-data_recipe.csv")
```

```python
def ingredients_to_text(ingredient_list):
    return " ".join(eval(ingredient_list))
```

```python
ingredients_to_text(data.ingredients[0])
```

```python
data.ingredients =data.ingredients.apply(ingredients_to_text)
```

```python
data.ingredients.sample().values
```

```python
data['ingredients'][:3].values.astype('U')
```

```python
count_vect = TfidfVectorizer(max_df=0.6, min_df=2, stop_words='english')
```

```python
doc_term_matrix =count_vect.fit_transform(data['ingredients'].values.astype('U'))
```

```python
LDA = LatentDirichletAllocation(n_components=8, random_state=42)
LDA.fit(doc_term_matrix)
```

```python
top_topic_words =LDA.components_[3].argsort()[-20:]
```

```python
for i in top_topic_words:
    print(count_vect.get_feature_names()[i])
```

```python
visual_sklearn = pyLDAvis.sklearn.prepare(LDA, doc_term_matrix, count_vect)
# Visualization can be displayed in the notebook
pyLDAvis.display(visual_sklearn)
```

```python
count_vect = TfidfVectorizer(max_df=0.6, min_df=2, stop_words='english')
doc_term_matrix =count_vect.fit_transform(data['ingredients'].values.astype('U'))
```

```python
X_test = count_vect.transform(['pitted fresh dark sweet cherries such as Bing bottled cherry juice almond extract brown sugar white sugar all-purpose flour pastry for a double-crust 9-inch pie butter'])
doc_topic_dist_unnormalized = np.matrix(LDA.transform(X_test))

# normalize the distribution (only needed if you want to work with the probabilities)
doc_topic_dist = doc_topic_dist_unnormalized/doc_topic_dist_unnormalized.sum(axis=1)
```
